```python
#import libraries

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
#read the test and train dataset
df_train=pd.read_csv("train.csv")
df_test=pd.read_csv("test.csv")
```

Dropping the column ID as it not required for model prediction

```python
df_train=df_train.drop(['ID'], axis=1)
df_test=df_test.drop(['ID'], axis=1)
```

### If for any column(s), the variance is equal to zero, then you need to remove those variable(s)

```python
var_col_zero=df_train.var()[df_train.var()==0].index.values
```

```python
#column index having zero variance
var_col_zero
```

```python
#dropping the column having zero variance
df_train1=df_train.drop(var_col_zero, axis=1)
df_test1=df_test.drop(var_col_zero, axis=1)
```

```python
#checking the initial columns and final columns after droping column having variance=0
df_train.shape[1], df_train1.shape[1]
```

```python
df_test.shape[1], df_test1.shape[1]
```

### Applying Label Encoder

```python
#finding the column having datatype as object
obj_col=df_train1.describe(include=[object]).columns.values
obj_col
```

```python
#label encoder
le=LabelEncoder()
for i in obj_col:
    le.fit(df_train1[i].append(df_test1[i]).values)
    df_train1[i]=le.transform(df_train1[i])
    df_test1[i]=le.transform(df_test1[i])
```

```python
#create X and y
X=df_train1.drop(['y'], axis=1)
y=df_train1['y']
```

```python
# create test and train split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=7)
```

### Performing dimensionality reduction

```python
from sklearn.decomposition import PCA
```

```python
#using PCA decomposition of the data from 364 components with 98% variance to lower dimension of 12 compnonents
pca=PCA(0.98, svd_solver='full')
sklearn_pca=pca.fit(X)
```

```python
sklearn_pca.n_components_
```

```python
sklearn_pca.explained_variance_ratio_
```

### predicting using XGBoost

```python
pca_xtrain=pd.DataFrame(pca.transform(xtrain))
pca_xtest=pd.DataFrame(pca.transform(xtest))
pca_df_test1=pd.DataFrame(pca.transform(df_test1))
```

```python
import xgboost as xgb
```

```python
xgb_reg=xgb.XGBRegressor(objective='reg:squarederror',learning_rate=0.1)
xgb_reg.fit(pca_xtrain,ytrain)
```

```python
xgb_reg_pred_ytest=xgb_reg.predict(pca_xtest)
```

```python
print(sqrt(mean_squared_error(ytest,xgb_reg_pred_ytest)))
```

```python
xgb_RFreg=xgb.XGBRFRegressor(objective='reg:squarederror',learning_rate=1)
xgb_RFreg.fit(pca_xtrain,ytrain)
```

```python
xgb_RFreg_pred_ytest=xgb_RFreg.predict(pca_xtest)
```

```python
print(sqrt(mean_squared_error(ytest,xgb_RFreg_pred_ytest)))
```
